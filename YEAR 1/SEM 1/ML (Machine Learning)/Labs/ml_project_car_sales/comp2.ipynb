```python
import pandas as pd

file_path = 'car_prices.csv'
data_car_prices = pd.read_csv(file_path)
data_car_prices.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix for the specified numerical columns
correlation_data = data_car_prices[['sellingprice', 'condition', 'odometer', 'year', 'mmr']].corr()

# Displaying a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', cbar=True)
plt.title("Correlation Heatmap between Selling Price and Selected Features")
plt.show()

# Scatter plots with regression lines for each pair with 'sellingprice'
features = ['condition', 'odometer', 'year', 'mmr']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data_car_prices, x=feature, y='sellingprice', alpha=0.3)
    sns.regplot(data=data_car_prices, x=feature, y='sellingprice', scatter=False, color='blue', line_kws={"lw":2})
    plt.title(f'Scatter Plot of Selling Price vs {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Selling Price')
    plt.show()

# Boxplot for condition and year to see the distribution of selling price across these categories
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_car_prices, x='condition', y='sellingprice')
plt.title("Boxplot of Selling Price by Condition")
plt.xlabel("Condition")
plt.ylabel("Selling Price")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=data_car_prices, x='year', y='sellingprice')
plt.title("Boxplot of Selling Price by Year")
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.xticks(rotation=45)
plt.show()
```
