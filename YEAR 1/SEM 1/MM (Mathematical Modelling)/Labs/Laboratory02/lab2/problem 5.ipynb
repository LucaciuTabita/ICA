```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Define the coefficients of the characteristic equation
A = np.array([[1, -5, 6]])
n_terms = 30  # Number of terms to compute

# Characteristic equation: r^2 - 5r + 6 = 0
r = np.roots([1, -5, 6])
print("Roots of the characteristic equation:", r)

# General solution based on the roots
x = np.zeros(n_terms)
x[0], x[1] = 1, 1  # Initial conditions x0 = 1, x1 = 1

# Compute further terms using the recurrence relation
for n in range(2, n_terms):
    x[n] = 5 * x[n-1] - 6 * x[n-2]

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(x, label='x_n', marker='o')
plt.title("Solution of the difference equation (a)")
plt.xlabel("n")
plt.ylabel("x_n")
plt.grid()
plt.legend()
plt.show()
```

```python
# (b) Define the coefficients for the homogeneous part
n_terms = 30
y_h = np.zeros(n_terms)  # Homogeneous solution
y_n = np.zeros(n_terms)   # Complete solution
y_n[0], y_n[1] = 1, 2     # Initial conditions y0 = 1, y1 = 2

# Solve the homogeneous part: y_n+2 - 3y_n+1 + 2y_n = 0
r_h = np.roots([1, -3, 2])
print("Roots of the homogeneous characteristic equation:", r_h)

# Homogeneous solution
for n in range(2, n_terms):
    y_h[n] = 3 * y_h[n-1] - 2 * y_h[n-2]

# Now find the particular solution
# Guess a polynomial solution of the form: An^2 + Bn + C
A, B, C = 1, 3, 2  # Coefficients for the particular solution
for n in range(2, n_terms):
    y_n[n] = 2*n**2 + 6*n + (3 * y_n[n-1] - 2 * y_n[n-2])

# Combine homogeneous and particular solutions
y_n += y_h

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(y_n, label='y_n', marker='o', color='orange')
plt.title("Solution of the difference equation (b)")
plt.xlabel("n")
plt.ylabel("y_n")
plt.grid()
plt.legend()
plt.show()
```

```python
# (c) Define the coefficients for the homogeneous part
n_terms = 30
y_h = np.zeros(n_terms)  # Homogeneous solution
y_n = np.zeros(n_terms)   # Complete solution
y_n[0], y_n[1] = 2, 1     # Initial conditions y0 = 2, y1 = 1

# Solve the homogeneous part: y_n+2 + 3y_n+1 + 2y_n = 0
r_h = np.roots([1, 3, 2])
print("Roots of the homogeneous characteristic equation:", r_h)

# Homogeneous solution
for n in range(2, n_terms):
    y_h[n] = -3 * y_h[n-1] - 2 * y_h[n-2]

# Now find the particular solution
# Guess a polynomial solution of the form: An^2 + Bn + C times 3^n
A, B, C = 1, 3, 2  # Coefficients for the particular solution
for n in range(2, n_terms):
    y_n[n] = 3**n * (2*n**2 + 4*n) + (-3 * y_n[n-1] - 2 * y_n[n-2])

# Combine homogeneous and particular solutions
y_n += y_h

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(y_n, label='y_n', marker='o', color='green')
plt.title("Solution of the difference equation (c)")
plt.xlabel("n")
plt.ylabel("y_n")
plt.grid()
plt.legend()
plt.show()
```
