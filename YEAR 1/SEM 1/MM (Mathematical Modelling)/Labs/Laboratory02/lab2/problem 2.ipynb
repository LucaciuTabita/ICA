```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve
```

```python
# Define the recurrence relation
def recurrence_a(a_n):
    return -a_n + 2

# Solve for the equilibrium point
a = symbols('a')
equilibrium_a = solve(Eq(a, recurrence_a(a)), a)
print("Equilibrium for (a):", equilibrium_a)

# Cobweb plot for (a)
def cobweb_plot_a(a_0, n_steps=20):
    x = np.linspace(-10, 10, 400)
    y = recurrence_a(x)
    plt.plot(x, y, 'b', label=r'$a_{n+1} = -a_n + 2$')
    plt.plot(x, x, 'r', label=r'$a_{n+1} = a_n$')

    a_n = a_0
    for _ in range(n_steps):
        a_next = recurrence_a(a_n)
        plt.plot([a_n, a_n], [a_n, a_next], 'k')
        plt.plot([a_n, a_next], [a_next, a_next], 'k')
        a_n = a_next

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xlabel(r'$a_n$')
    plt.ylabel(r'$a_{n+1}$')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot for different initial values
initial_values = [0, 1, -1, 2, -2]
for a_0 in initial_values:
    print(f"\nCobweb plot for initial value a_0 = {a_0}")
    cobweb_plot_a(a_0)
```
